```python
# We can see that there also ranges in the "total_sqft" column. We will replace that range with with the mean.
```

```python
def convertrange(x):
    temp = x.split("-")
    if len(temp) == 2:
        return ((float(temp[0]) + float(temp[1])) /2 )
    try:
        return float(x)
    except:
        return None
```

```python
df["total_sqft"] = df["total_sqft"].apply(convertrange)
```

## Outlier Detection and removal

```python
df.describe()
```

```python
(df["total_sqft"] / df["bhk"]).describe()
```

```python
df = df[(df["total_sqft"] / df["bhk"]) >= 300]
df
```

```python
df.describe()
```

```python
df.info()
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    for key, sub_df in df.groupby("location"):
        mean = np.mean(sub_df.price_per_sqft)
        std = np.std(sub_df.price_per_sqft)
        
        # Only keep those prices_per_sqft which are within one standard deviation away from mean.
        gen_df = sub_df[((sub_df.price_per_sqft) > (mean - std)) & ((sub_df.price_per_sqft) <= (mean + std))]
        df_output = pd.concat([df_output, gen_df], ignore_index = False)

    return df_output

df = remove_outliers_sqft(df)
df
```

```python
df.describe()
```

```python
def remove_outlier_bhk(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby("location"):
        bhk_stats = {}

        for bhk, bhk_df in df.groupby("bhk"):
            bhk_stats[bhk] = {
                "mean" : np.mean(bhk_df.price_per_sqft),
                "std" : np.std(bhk_df.price_per_sqft),
                "count" : bhk_df.shape[0]
            }
        
    for bhk, bhk_df in location_df.groupby("bhk"):
        stats = bhk_stats.get(bhk - 1)
        if stats and stats["count"] > 5:
            exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < stats["mean"]].index.values)

    return df.drop(exclude_indices, axis = "index")
```

```python
df = remove_outlier_bhk(df)
```

## Cleaned Data

```python
df.head()
```

```python
df.info()
```

```python
df["bath"] = df["bath"].apply(lambda x: np.random.randint(1, 6) if np.isnan(x) else x)
```

```python
X = df.drop(columns = ["price"])
y = df["price"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
print(X_train.shape)
print(X_test.shape)
```

## Applying Linear Regression

```python
column_trans = make_column_transformer((OneHotEncoder(sparse_output = False), ["location"]),
                                      remainder = "passthrough")
```

```python
scaler = StandardScaler()
lr = LinearRegression()
pipe = make_pipeline(column_trans, scaler, lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
r2_score(y_test, y_pred_lr)
```

## Lasso

```python
lasso = Lasso()
pipe = make_pipeline(column_trans, scaler, lasso)
pipe.fit(X_train, y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test, y_pred_lasso)
```

## Ridge

```python
ridge = Ridge()
pipe = make_pipeline(column_trans, scaler, ridge)
pipe.fit(X_train, y_train)
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test, y_pred_ridge)
```

```python
print("No Regularization: ", r2_score(y_test, y_pred_lr))
print("Lasso: ",r2_score(y_test, y_pred_lasso))
print("Ridge: ", r2_score(y_test, y_pred_ridge))
```
