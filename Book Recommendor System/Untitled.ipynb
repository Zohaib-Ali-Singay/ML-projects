Why the "number_ratings" is not displayed ?

The issue is with the `inplace=True` argument used in the `rename` method. When `inplace=True` is used, the `rename` method modifies the DataFrame in place and returns `None`. As a result, `number_ratings` is set to `None`.

To resolve this, you can either remove `inplace=True` and capture the renamed DataFrame in a variable, or perform the renaming in a separate step. Here's how to do it both ways:

### Method 1: Capture the Renamed DataFrame in a Variable
```python
number_ratings = ratings_with_books.groupby("title")["rating"].count().reset_index()
number_ratings = number_ratings.rename(columns={"rating": "No of ratings"})
print(number_ratings)
```

### Method 2: Perform Renaming in a Separate Step
```python
number_ratings = ratings_with_books.groupby("title")["rating"].count().reset_index()
number_ratings.rename(columns={"rating": "No of ratings"}, inplace=True)
print(number_ratings)
```

### Explanation:
- **Grouping and Counting**: `ratings_with_books.groupby("title")["rating"].count()` groups the DataFrame by the `title` column and counts the number of ratings for each title.
- **Reset Index**: `.reset_index()` resets the index of the resulting DataFrame, converting the group keys into columns.
- **Renaming Columns**: `.rename(columns={"rating": "No of ratings"}, inplace=True)` renames the `rating` column to `No of ratings`.

By following either of these methods, you ensure that `number_ratings` holds the modified DataFrame, and the DataFrame is displayed correctly when printed.

````python
final_df = ratings_with_books.merge(number_ratings, on = "title") # We are doing it so that no of ratings is also displayed along with each book.
final_df
````

````python
final_rating = final_df[final_df["No of ratings"] >= 50] # Filters those books which have ratings greater than equal to 50.
final_rating
````

````python
final_rating.shape # final_rating contains only those books which have ratings greater than or equal to 50 and only those users which have given more 
# 200 ratings.
````

````python
final_rating.drop_duplicates(subset = ["user_id", "title"], inplace = True)
````

````python
book_pivot = final_rating.pivot_table(columns = "user_id", index = "title", values = "rating")
book_pivot
````

````python
book_pivot.shape # Some users which give more than 200 ratings, rated those books which were having rating less than 50. So those users are dropped.
````

````python
book_pivot.fillna(0, inplace = True)
````

````python
book_pivot.astype(int)
````

````python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot) # It will keep only the non-zero values from the above sparse matrix
````

````python
book_sparse
````

````python
book_sparse.shape
````

````python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = "brute")
model.fit(book_sparse) # We will keep n_neighbors = 5 by default. We can change n_neighbors but we will work with its default value. Changing
# n_neighbors can change the performance.
````

````python
distances, suggestions = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1, -1), n_neighbors = 6)
````

````python
distances # 0 is the distance of book with itself. Therefore, we have taken n_neighbors = 6 here.
````

````python
suggestions
````

````python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
````

````python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distances, suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors = 6)
    suggestions = np.array(suggestions)
    suggestions = suggestions.flatten()
    for i in range(len(suggestions)):
        if i!=0:
            print(book_pivot.index[suggestions[i]])
````

````python
recommend_book("Animal Farm")
````
