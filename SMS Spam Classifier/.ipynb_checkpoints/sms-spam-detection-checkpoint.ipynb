## 3. Data Preprocessing for textual data.  
The following steps are generally followed for preprocessing textual data.  
- Lower Case
- Tokenization
- Removing Special Characters
- Removing stop words and punctuation
- Stemming or Lemmaization

```python
from nltk.corpus import stopwords
import string
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
ps.stem("loving")
```

```python
stopwords.words("english")
```

```python
string.punctuation
```

```python
def transform_text(text):
    # Lower Case
    text = text.lower()

    # Tokenization
    text = nltk.word_tokenize(text)

    # Removing Special Characters
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)

    text = y.copy()
    y.clear()

    # Removing Stop Words and Punctuation
    for i in text:
        if i not in stopwords.words("english") and i not in string.punctuation:
            y.append(i)

    text = y.copy()
    y.clear()
    # Stemming
    for i in text:
        y.append(ps.stem(i))    
    
    return " ".join(y)
```

```python
transform_text("So, what's up you all. How's your day passing. Are you dancing and singing")
```

### CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
cv = CountVectorizer()
Xcv = cv.fit_transform(df["transformed_text"]).toarray()
```

```python
Xcv.shape
```

```python
y = df["Target"].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_traincv, X_testcv, y_train, y_test = train_test_split(Xcv, y, test_size = 0.2, random_state = 22)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
```

```python
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

```python
gnb.fit(X_traincv, y_train)
y_pred1 = gnb.predict(X_testcv)
print(accuracy_score(y_test, y_pred1))
print(precision_score(y_test, y_pred1))
print(recall_score(y_test, y_pred1))
sns.heatmap(confusion_matrix(y_test, y_pred1), annot = True, fmt = ".2f")
```

```python
mnb.fit(X_traincv, y_train)
y_pred2 = mnb.predict(X_testcv)
print(accuracy_score(y_test, y_pred2))
print(precision_score(y_test, y_pred2))
print(recall_score(y_test, y_pred2))
sns.heatmap(confusion_matrix(y_test, y_pred2), annot = True, fmt = ".2f")
```

```python
bnb.fit(X_traincv, y_train)
y_pred3 = bnb.predict(X_testcv)
print(accuracy_score(y_test, y_pred3))
print(precision_score(y_test, y_pred3))
print(recall_score(y_test, y_pred1))
sns.heatmap(confusion_matrix(y_test, y_pred3), annot = True, fmt = ".2f")
```

### TfidfVectorizer

```python
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

```python
gnb.fit(X_traintf, y_train)
y_pred1 = gnb.predict(X_testtf)
print(accuracy_score(y_test, y_pred1))
print(precision_score(y_test, y_pred1))
print(recall_score(y_test, y_pred1))
sns.heatmap(confusion_matrix(y_test, y_pred1), annot = True, fmt = ".2f")
```

```python
mnb.fit(X_traintf, y_train)
y_pred2 = mnb.predict(X_testtf)
print(accuracy_score(y_test, y_pred2))
print(precision_score(y_test, y_pred2))
print(recall_score(y_test, y_pred2))
sns.heatmap(confusion_matrix(y_test, y_pred2), annot = True, fmt = ".2f")
```

```python
bnb.fit(X_traintf, y_train)
y_pred3 = bnb.predict(X_testtf)
print(accuracy_score(y_test, y_pred3))
print(precision_score(y_test, y_pred3))
print(recall_score(y_test, y_pred1))
sns.heatmap(confusion_matrix(y_test, y_pred3), annot = True, fmt = ".2f")
```

### Comparison of Multinomial Naive Bayes with other models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, ExtraTreesClassifier, GradientBoostingClassifier, BaggingClassifier
from xgboost import XGBClassifier
```

```python
svc = SVC(kernel = "linear", gamma = 1.0)
knc = KNeighborsClassifier()
mnb = MultinomialNB()
dtc = DecisionTreeClassifier(max_depth = 5)
lrc = LogisticRegression(solver = "liblinear", penalty = "l1")
rfc = RandomForestClassifier(n_estimators = 50, random_state = 42)
abc = AdaBoostClassifier(n_estimators = 50,random_state = 42)
bc = BaggingClassifier(n_estimators = 50, random_state = 42)
etc = ExtraTreesClassifier(n_estimators = 50, random_state = 42)
gbc = GradientBoostingClassifier(n_estimators = 50, random_state = 42)
xgb = XGBClassifier(n_estimators = 50, random_state = 42)
```

```python
clfs = {
    "svc" : svc,
    "knc" : knc,
    "mnb" : mnb,
    "dtc" : dtc,
    "lrc" : lrc,
    "rfc" : rfc,
    "abc" : abc,
    "etc" : etc,
    "gbc" : gbc,
    "bc" : bc,
    "xgb" : xgb
}
```

```python
def train_classifier(clf):
    clf.fit(X_traintf, y_train)
    y_pred = clf.predict(X_testtf)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)

    return accuracy, precision
```

```python
accuracy_scores = []
precision_scores = []

for name, clf in clfs.items():
    current_accuracy, current_precision = train_classifier(clf)

    print("For", clf)
    print("Accuracy Score:", current_accuracy)
    print("Precision Score:", current_precision)

    accuracy_scores.append(current_accuracy)
    precision_scores.append(current_precision)
    
```

```python
performance_df = pd.DataFrame({"Algorithms" : clfs.keys(), "Accuracy Score" : accuracy_scores, "Precision Scores" : [0.9767441860465116,
 1.0,
 0.9906542056074766,
 0.8145161290322581,
 0.865546218487395,
 0.9913793103448276,
 0.8421052631578947,
 0.975609756097561,
 0.9611650485436893,
 0.8897058823529411,
 0.968503937007874]}).sort_values("Precision Scores", ascending = False)
```

```python
performance_df
```

```python
# We see that that precision of knc is highest but its accuracy score is low. rfc'c precision is a bit low but its accuracy is high as compared to knc.
# Therefore, we will choose RandomForestClassifier for balanced results.
```
