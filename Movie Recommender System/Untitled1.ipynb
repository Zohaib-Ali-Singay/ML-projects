```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies["genres"].apply(convert)
```

```python
# Thus we have got the desired format by "convert and literal_eval funtion".
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
# We will do the same with keywords column
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
# Now we will select top 3 actors from each "cast" column. But lets see what is actually in the cast column by examining only its single cell.
```

```python
movies["cast"][0]
```

```python
# each dictionary in the above list is giving us info about each actor. We will select only first 3 dictionary for each cell of "cast".
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i["name"])
            counter = counter + 1
        else:
            break
    return L
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast'] = movies['cast'].apply(convert3) 
```

```python
# Now only "crew" is left
```

```python
movies['crew'][0]
```

```python
# We will fetch the "name" whose job is director i.e, in "crew" column, we will keep only the director's name.
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            L.append(i['name'])
            break
    return L
```

```python
movies["crew"].apply(fetch_director)
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
# We see that each cell in "overview" column is a string. We have to convert it into "list" type so that it is compatible with "genres", "keywords", 
#"cast" and "crew" column
```

```python
movies["overview"].apply(lambda x: x.split())
```

```python
movies["overview"] = movies["overview"].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
# Now we will finish the space in each words of each list of each column. Why ?
# Because see the above df. We can see that in first row of "cast", "Sam Worthington" is present and in the 3rd row of "crew", "Sam Mendes" is present.
# If we keep theses spaces as it is, then our recommender system will be confused that on the basis of which "Sam" the movies should be recommended.
# Therefore we will finish the spaces in each word and convert them into a single word.

# We will remove spaces in each word of "genres", "keywords", "cast", "crew" column.
```

```python
movies["genres"].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies["genres"] = movies["genres"].apply(lambda x: [i.replace(" ", "") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x: [i.replace(" ", "") for i in x])
movies["cast"] = movies["cast"].apply(lambda x: [i.replace(" ", "") for i in x])
movies["crew"] = movies["crew"].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
# See the above array. There are many similar words such as [abilities, ability], [activities, activity], ['accept', 'accepted', 'accept'] and 
# [accidental, accidentally] etc, we have to convert them into single words like ability, activity, accept, accident etc respectively. This process is
# called stemming. We will do it through "nltk" library, a famous NLP library. An example is illustrated below.
```

```python
[["love", "lover", "loving"]]
# The above array will be converted to [['love', 'love', 'love']] with the help of stemming. Now we will apply stemming directly on new_df["tags"].
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem("accidentally")
```

```python
def stem_func(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df["tags"][0]
```

```python
stem_func(new_df["tags"][0])
```

```python
new_df["tags"] = new_df["tags"].apply(stem_func)
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
# Now we see that the similar words are replaced by a single word.
```

```python
# We will not calculate euclidean distance in calculating the distance between two movies i.e, two vectors, it is because in higher dimensions (5000 in this case)
# , euclidean distance fails. Thus we will use another distance called cosine distance which measures the angle between the vectors to calculate how 
# much they are close to or far away from each other. The more the vectors are close to each other, the more similar are the movies and vice versa.
# Distance is inversely proportional to similarity.
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape # We have calculated the similarity scores of each movie with every movie.
```

```python
similarity[0] # It is telling the similarity of 1st movie with ever movie
```

```python
new_df[new_df["title"] == "Batman Begins"].index[0] # It will give us the actual index of the movie in the dataframe and we will make reommendations 
# on the basis of index i.e, when I give index (or the movie name), then which movie(s) is/are similar to this movie, will be recommended.
# These recommendations will be based on cosine similarity score we just calculated above.Here in this project, we will recommend the 5 most closest 
# movies when we are given a particular movie name.
```

```python
def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity[movie_index]
    return distances
```

```python
recommend("Batman Begins")
```

```python
# Now we will sort this array in descending order but the problem here is this, that if we sort this array in descending order, we will lose the
# original index order (or movie order) i.e, the cosine distances will be shuffled. Therefore, we will have to shuffle the array in such a way that 
# the original index position is also reserved in the sorted array.
```

```python
sorted(list(enumerate(similarity[0])))
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x: x[1]) # This will do the sorting on the basis of 2nd element of each tuple in
# the list and that 2nd element is actually the similarity score
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x: x[1]) [1 : 6]# Sort the top 5 closest movies 
```

```python
# Thus we will write the above line of code in the "recommend" function.
```

```python
def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x: x[1])[1:6]

    for i in movies_list:
        print(i[0])
```

```python
recommend("Avatar")
```
